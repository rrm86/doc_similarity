```python
import pickle
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
with open('cosine_distance', 'rb') as f:
    cosine_distance = pickle.load(f)
    
with open('wmd', 'rb') as f:
    wmd = pickle.load(f)
```

```python
cosine_distance_pd = pd.DataFrame(cosine_distance)
cosine_distance_pd = cosine_distance_pd.iloc[10:22, 10:22]
```

```python
f,ax = plt.subplots(figsize=(19, 19))
sns.heatmap(cosine_distance_pd, annot=True, linewidths=.4, fmt= '.2f',ax=ax, cmap="Greens")
plt.show()
```

```python
wmd_pd = pd.DataFrame(wmd)
wmd_pd = wmd_pd.iloc[10:22, 10:22]
```

```python
f,ax = plt.subplots(figsize=(19, 19))
sns.heatmap(wmd_pd, annot=True, linewidths=.4, fmt= '.2f',ax=ax,cmap="Blues")
plt.show()
```
